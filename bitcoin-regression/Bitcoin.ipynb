```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import sklearn
import re
```

```python
df=pd.read_csv('Bitcoin.csv')
```

```python
 #Converting the time span 
```

```python
df['Timestamp'] = pd.to_datetime(df['Timestamp'], unit='s')
```

```python
#Selection of Features
```

```python
X = df[['Open', 'High', 'Low', 'Volume']]
y = df['Close']
```

```python
#Splitting of the dataset 
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.3,
    random_state=42
)
```

```python
#Performing feature scalling
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
#Importing the model (Liner Regression)
```

```python
from sklearn.linear_model import LinearRegression

lr_model = LinearRegression()
lr_model.fit(X_train, y_train)

lr_pred = lr_model.predict(X_test)
```

```python
#Importing the model (K nearest neighbors regression)
```

```python
from sklearn.neighbors import KNeighborsRegressor

knn_model = KNeighborsRegressor(n_neighbors=5)
knn_model.fit(X_train_scaled, y_train)

knn_pred = knn_model.predict(X_test_scaled)
```

```python
#Performing evaluation metrics
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
#Linear Regression result
```

```python
print("===== LINEAR REGRESSION =====")
print("MAE:", mean_absolute_error(y_test, lr_pred))
print("MSE:", mean_squared_error(y_test, lr_pred))
print("R2 Score:", r2_score(y_test, lr_pred))
```

```python
print("===== KNN REGRESSION =====")
print("MAE:", mean_absolute_error(y_test, knn_pred))
print("MSE:", mean_squared_error(y_test, knn_pred))
print("R2 Score:", r2_score(y_test, knn_pred))
```

## Model Evaluation
Linear Regression performed better overall, because it has lower MAE and MSE, meaning its predictions are closer to the actual values and more consistent than KNN.
Both models are extremely accurate (R² ≈ 1 i.e, our model is almost perfectly predicting the actual values with very little error), showing they explain almost all the variation in the data, but Linear Regression is slightly more reliable based on error values.
